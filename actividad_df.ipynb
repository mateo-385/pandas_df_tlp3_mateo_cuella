```python
import pandas as pd
import sqlite3

#Aqui agrega la conexión a la base de datos
conn = sqlite3.connect("Salaries.sqlite")

#Crea el dataframe con los datos del datasets
df = pd.read_sql_query("SELECT * FROM salaries", conn)
conn.close()

df.head()
```

1. **Top 10 empleados con mayor salario total (incluyendo beneficios)**
Extrae e imprime un top 10 de expleados con mayor salario incluyendo los beneficios 


```python
#todo

# se ordenan los valores de la columna "TotalPayBenefits" de manera descendiente e ignorando la columna del index
# solo se muestran las tablas de "EmployeeName" y "TotalPayBenefits" para una mejor lectura de los resultados
# se limitan los resultados a los primeros 10 con ".head(10)"

df.sort_values("TotalPayBenefits", ascending=False, ignore_index=True)[['EmployeeName', 'TotalPayBenefits']].head(10)
```

2. **Filtrar empleados con más de 50,000 en horas extra, la columna a tratar es: (OvertimePay)**

```python
#todo

# se convirte la columna a numerica (los errores se vuelven NaN)
df["horas_extras"] = pd.to_numeric(df["OvertimePay"], errors="coerce")

# se muestran los resultados mayores a 50,0000
df[df["horas_extras"] > 50000]
```
